```python
import tensorflow as tf

import matplotlib.pyplot as plt
import numpy as np
import os
from tensorflow.keras import layers,models
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
image_size=(180,180)
batch_size=32
```

```python
train_data=tf.keras.preprocessing.image_dataset_from_directory(
    '/content/chest_xray/chest_xray/train',image_size=image_size,batch_size=batch_size
)
val_data=tf.keras.preprocessing.image_dataset_from_directory(
    '/content/chest_xray/chest_xray/val',image_size=image_size,batch_size=batch_size
)
test_data=tf.keras.preprocessing.image_dataset_from_directory(
    '/content/chest_xray/chest_xray/test',image_size=image_size,batch_size=batch_size,shuffle=False

)
```

```python
class_names = train_data.class_names
print(":", class_names)
```

```python
model=models.Sequential([
    layers.Rescaling(1./255,input_shape=(180,180,3)),
    layers.Conv2D(16,3,activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(32,3,activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(64,3,activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history=model.fit(train_data,validation_data=val_data,epochs=3)
```

```python
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title("    ")
plt.show()
```

```python
y_true = []
y_pred = []

for images, labels in test_data:
    preds = model.predict(images)
    preds = tf.squeeze(tf.round(preds)).numpy()
    y_pred.extend(preds)
    y_true.extend(labels.numpy())

#  
print(" Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

#  
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("  ")
plt.show()
```
